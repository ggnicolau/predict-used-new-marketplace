```python
import pandas as pd
df_raw = pd.read_json('MLA_100k_checked_v3.jsonlines', lines=True)
df_backup = df_raw.head(1000)
```

```python
df = df_backup
```

```python
df['seller_country'] = df.apply(lambda x : x['seller_address']['country']['name'], axis = 1)
df['seller_state'] = df.apply(lambda x : x['seller_address']['state']['name'], axis = 1)
df['seller_city'] = df.apply(lambda x : x['seller_address']['city']['name'], axis = 1)
```

```python
import ast
def str_to_dict(column):
    for i in range(len(column)):
        try:
            print(i)
            column[i] = ast.literal_eval(column[i][0])
        except:
            column[i] = column[i]

str_to_dict(df['descriptions'])
```

```python
df = pd.concat([df, df["descriptions"].apply(pd.Series)], axis=1)
df = pd.concat([df, df["shipping"].apply(pd.Series)], axis=1)
```
